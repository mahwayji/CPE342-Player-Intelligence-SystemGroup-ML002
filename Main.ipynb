# This is the answer creating notebook

```python
import pandas as pd 
import numpy as np
import joblib
```

## Prepare the file

```python
df = pd.DataFrame()

# get the size of the test
test1 = pd.read_csv('task1/test.csv')

df['id'] = test1['id']
df
```

## Task1

## Modify the dataset to match the training set

```python
test1.drop(['id', 'player_id'], axis = 1, inplace = True)
```

```python
model1 = joblib.load('task1/model.pkl')

y_pred = model1.predict(test1)
y_pred
```

```python
df['task1'] = y_pred
df
```

# Task 2

```python
answer2 = pd.read_csv('task2/predicted.csv')

answer2.head()
```

```python
df['task2'] = [i for i in answer2['0']]

df
```

# Task 3

```python
test3 = pd.read_csv('task3/test.csv')
```

```python
test3.drop(['id', 'player_id'], axis = 1, inplace = True)
```

```python
test3.fillna(test3.mean(), inplace = True)
```

```python
model3 = joblib.load('task3/model3.pkl')

y_pred = model3.predict(test3)

y_pred = np.maximum(y_pred, 0)
y_pred
```

```python
df['task3'] = y_pred

df
```

# Task 4

```python
df['task4'] = [-1.0 for i in range(len(df))]

df
```

# Task 5

```python
answer5 = pd.read_csv('task5/predicted.csv')

answer5.head()
```

```python
df['task5'] = [1.0 if (i==-1) else 0.0 for i in answer5['0']]

df
```
