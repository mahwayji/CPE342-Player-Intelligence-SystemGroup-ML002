# Task 4

```python
test4 = pd.read_csv('task4/test.csv')

test4.head()
```

```python
from keras.models import load_model
from keras.preprocessing.image import load_img

model4 = load_model('task4/model4.h5')

image = load_img('task4/test/s0a8c3o.jpg', target_size = (128,128))
img = np.array(image)
img = img/255.0
img = img.reshape(1,128,128,3)

proba = model4.predict(img)
label = np.argmax(proba, axis=-1)   
print(label)
```

```python
import os
default_path = os.path.join('task4', 'test')

y_pred = []
for i in test4['file_name']:
    image = load_img(os.path.join(default_path, i), target_size = (128,128))
    img = np.array(image)
    img = img/255.0
    img = img.reshape(1,128,128,3)
    proba = model4.predict(img, verbose = 0)
    labeled = np.argmax(proba[0], axis=-1)
    y_pred.append(labeled)    
    
y_pred
```

```python
df['task4'] = y_pred

df
```
